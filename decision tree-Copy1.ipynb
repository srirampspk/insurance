```python
df = df.drop("Unnamed: 0", axis=1, errors='ignore')
df
```

```python
from sklearn.preprocessing import LabelEncoder
ml=LabelEncoder()
```

```python
df['employee']=ml.fit_transform(df['Employment Type'])
```

```python
df=df.drop('Employment Type',axis=1,errors='ignore')
```

```python
df['GraduateOrNot']=df['GraduateOrNot'].replace({'Yes':1,'No':0})
df['FrequentFlyer']=df['FrequentFlyer'].replace({'Yes':1,'No':0})
df['EverTravelledAbroad']=df['EverTravelledAbroad'].replace({'Yes':1,'No':0})
df
```

```python
from sklearn.model_selection import train_test_split
```

```python
x=df[['Age','GraduateOrNot','AnnualIncome','FamilyMembers','ChronicDiseases','FrequentFlyer','EverTravelledAbroad','employee']]
```

```python
y=df['TravelInsurance']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

```python
from sklearn import tree
dec=tree.DecisionTreeClassifier()
```

```python
dec.fit(x_train,y_train)
```

```python
dec.score(x_test,y_test)
```
